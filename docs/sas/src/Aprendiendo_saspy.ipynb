```python
import saspy
saspy
```

```python
sas = saspy.SASsession()
sas
```

Más información en la [documentación de saspy](https://sassoftware.github.io/saspy/getting-started.html)

Cargar datos existentes, en este caso `class` de la librería `sashelp`.

```python
datos = sas.sasdata("class","sashelp")
```

Explorando la información

```python
type(datos)
```

```python
datos.columnInfo()
```

Usando los *SAS MAGIC* para ejecutar código SAS

```python
%%SAS
proc print data=sashelp.class;
run;

data work.carros;
  set sashelp.cars;
run;
```

Para salir de la sesión de SAS:

```python
sas.disconnect()
```
